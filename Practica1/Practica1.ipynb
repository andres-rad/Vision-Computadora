## 1. Implementar las siguientes funciones:

#### a) Suma, resta y producto de imágenes.

```python
def add(img1, img2):
    """
    params:
        img1, img2 --> imagenes a sumar
        
    output:
        _ --> Imagen resultante de sumar los inputs pixel a pixel
    """
    img1_ar = np.asarray(img1)
    img2_ar = np.asarray(img2)
    return Image.fromarray(img1_ar + img2_ar)

def substract(img1, img2):
    """
    params:
        img1, img2 --> imagenes a sumar
        
    output:
        _ --> Imagen resultante de restar los inputs pixel a pixel
    """
    img1_ar = np.asarray(img1)
    img2_ar = np.asarray(img2)
    return Image.fromarray(img1_ar - img2_ar)


def product(img1, img2):
    """
    params:
        img1, img2 --> imagenes a sumar
        
    output:
        _ --> Imagen resultante de multiplocar los inputs pixel a pixel
    """    
    img1_ar = np.asarray(img1)
    img2_ar = np.asarray(img2)
    return Image.fromarray(img1_ar * img2_ar)
```

```python
img = Image.open('img/Images_test/peppers256.png')

fig = plt.figure(figsize=(10,8))
plt.imshow(img)
plt.title("Imagen original", size=18)
plt.show()

fig, (ax, ax2, ax3) = plt.subplots(ncols=3, figsize=(25,25))

ax.imshow(add(img, img))
ax.set_title("Imagen sumada con sigo misma", size=18)

ax2.imshow(substract(img, img))
ax2.set_title("Imagen restada con sigo misma", size=18)


ax3.imshow(product(img, img))
ax3.set_title("Imagen multiplicada con sigo misma", size=18)

plt.show()
```

#### b) Producto de una imagen por un escalar.

```python
def scalar_product(img, n):
    """
    params:
        img1 --> imagen
        n --> escalar
    output:
        _ --> Imagen resultante de multiplicar los píxeles de la imagen por n
    """
    return img.point(lambda i: i * n)
```

```python
img = Image.open('img/Images_test/house.png')

fig, (ax, ax2) = plt.subplots(ncols=2)

ax.imshow(img)
ax.set_title("Imagen original", size=18)

ax2.imshow(scalar_product(img, 2))
ax2.set_title("Imagen multiplicada por escalar 2", size=18)

plt.show()
```

#### c) Compresión del rango dinámico: sean r el nivel de gris de la imagen de entrada y s el nivel de gris de la imagen de salida, tal que s = f(r), r en [0, R]. Entonces la función de compresión del rango dinámico es f(r) = c * log(r + 1), eligiendo c de manera tal que f(r) en [0, 255].

```python
def drc(img):
    """
    params:
        img --> imagen a la que se le aplica el filtro
    output:
        _ --> devuelve la imagen después de que se le aplica el filtro de compresion del rango dinámico
    """
    _, R = img.getextrema() # devuelve el rango de los valores de la imagen
    c = 254.0 / log(1 + R)
    return img.point(lambda i: int(c * log(1 + i)))
```

```python
img = Image.open('img/Images_test/barbara.png')

fig, (ax, ax2) = plt.subplots(ncols=2)

ax.imshow(img)
ax.set_title("Imagen original", size=12)

ax2.imshow(drc(img))
ax2.set_title("Imagen luego de aplicarle compresión del rango dinámico", size=12)

plt.show()
```

### 2) Implementar una función que devuelva el negativo de una imagen.

```python
def negate(img):
    _, R = img.getextrema()
    return img.point(lambda i: R - 1 - i)
```

### 3) Implementar una función que aplique un umbral a una imagen, devolviendo una imagen binaria.

```python
def threshold(img, treshold):
    return img.point(lambda i: 0 if i < treshold else 255).convert('1')
```

### 4) Fraccionamiento de los planos de bits: hacer un programa que separe una imagen en los 8 planos de bits y mostrarlos cada uno por separado.

```python
def bitplane(img, position):
    """
    params:
        img --> imagen
        position --> numero de bit que miro para cada valor del pixel
    
    output:
        _ --> imagen resultante de mirar solamente el bit en la posicion position
              para cada pixel de la imagen original
    """
    bit_masks = [0b00000001,
                 0b00000010,
                 0b00000100,
                 0b00001000,
                 0b00010000,
                 0b00100000,
                 0b01000000,
                 0b10000000
                ]
    return img.point(lambda i: 255 if i & bit_masks[position] else 0)

def bitplanes(img):
    """
    params:
        img --> imagen
    output:
        planes --> arreglo que tiene en la i-esima posicion la imagen resultante de mirar
                   solamente el bit i de la imagen original
    """
    planes = []
    for pos in range(8):
        planes.append(bitplane(img, pos))
    return planes

def show_bitplanes(img):
    """
    Funcion auxiliar para mostrar todos los bitplanes
    """
    planes = bitplanes(img)
    fig, axes = plt.subplots(2, 4, figsize=(30,30), sharex=True, sharey=True)
    for idx, plane in enumerate(planes):
        ax = axes[idx//4, idx%4]
        ax.set_title('Bit {}'.format(idx), size=30)
        ax.imshow(planes[idx])
    return fig
```

```python
img = Image.open(img_path + 'Images_test/flinstones.png')

fig = plt.figure(figsize=(10, 8))
plt.title('Imagen Original', size=18)
plt.imshow(img)
plt.show()

show_bitplanes(img)
plt.show()
```

#### 5) Implementar una función que devuelva el histograma de niveles de gris de una imagen.

```python
def histogram(img):
    """
    params:
        img --> imagen de entrada
    
    output:
        amts --> diccionario que tiene para cada valor la cantidad de veces que aparece
    """
    amts = { i: 0 for i in range(256)}
    for pixel_value in list(img.getdata()):
        amts[pixel_value] += 1

    return amts
```

```python
im = Image.open(img_path + 'ImagenesHistograma/Fig30.jpg')
hist = histogram(im)

fig, (ax, ax2) = plt.subplots(ncols=2)

ax.imshow(im)
ax2.bar(list(hist.keys()), list(hist.values()))

ax.set_title("Imagen original")
ax2.set_title("Histograma original")

asp = np.abs(np.diff(ax2.get_xlim())[0] / np.diff(ax2.get_ylim())[0])
asp /= np.abs(np.diff(ax.get_xlim())[0] / np.diff(ax.get_ylim())[0])

ax2.set_aspect(asp)

plt.show()
```

#### 6) Examinando el histograma del ejercicio anterior, implementar una función que devuelva una imagen que tenga aumento del contraste.

```python
def augment_contrast(img, r1, r2, s1, s2):
    """
    params:
        img --> imagen de entrada
        r1, r2, s1, s2 --> enteros que indican los intervalos 
                           a los que se mapean los colores
    output:
        _ --> imagen modificada por el filtro
        
    Se crean tres ecuaciones lineales que f1, f2, f3 con dominios:
        f1 : [0, r1) --> [0, s1)
        f2 : (r1, r2) --> (s1, s2)
        f3 : [r2, 255] --> [s2, 255]
        
    Para cada pixel se lo transforma con la función correspondiente a su valor
    """
    def map_color(pixel_value):
        if pixel_value < r1:
            m = s1 / r1
            return int(m * pixel_value)
        
        elif pixel_value < r2:
            m = (s2 - s1) / (r2 - r1)
            b = s1 - m * r1
            return int(m * pixel_value + b)
        
        else:
            m = (255 - s2) / (255 - r2)
            b = s2 - m * r2
            return int (m * pixel_value + b)
        
    return img.point(lambda x: map_color(x))

# dejo esto afuera para poder mostrar la función usada
def map_color(pixel_value, r1, r2, s1, s2):
    if pixel_value < r1:
        m = s1 / r1

        return int(m * pixel_value)
    elif pixel_value <= r2:
        m = (s2 - s1) / (r2 - r1)
        b = s1 - m * r1
        return int(m * pixel_value + b)
    else:
        m = (255 - s2) / (255 - r2)
        b = s2 - m * r2
        return int (m * pixel_value + b)
```

```python
im = Image.open(img_path + 'ImagenesHistograma/Fig30.jpg')

r1 = 50
r2 = 200
s1 = 10
s2 = 240

fig, (ax, ax2, ax3) = plt.subplots(ncols=3)

ax.imshow(im)
ax2.imshow(augment_contrast(im, r1, r2, s1, s2))
ax3.plot(list(range(256)), [map_color(x, 50, 200, 10, 240) for x in range(256)])
ax3.plot([r1, r2], [s1, s2], 'o')
ax3.vlines([r1, r2], 0, [s1, s2], linestyle="dashed", linewidth=0.8)
ax3.hlines([s1, s2], 0, [r1, r2], linestyle="dashed", linewidth=0.8)

ax.set_title("Imagen original")
ax2.set_title("Imagen con contraste aumentado")
ax3.set_title("Función de contraste utilizada")

asp = np.abs(np.diff(ax3.get_xlim())[0] / np.diff(ax3.get_ylim())[0])
asp /= np.abs(np.diff(ax.get_xlim())[0] / np.diff(ax.get_ylim())[0])

ax.set_aspect(asp)
ax2.set_aspect(asp)
ax3.set_aspect(asp)

plt.show()
```

#### 7) Implementar una función que resuelva la ecualización del histograma.

```python
def accumulated_histogram(img):
    """
    params:
        img --> imagen a la que se le calcula el histograma
        
    output:
        accumulated_hist --> es el histograma que aproxima la funcion de densidad
                             del valor de un pixel en la imagen pasada por input
    """
    hist = histogram(img)
    total = float(im.size[0] * im.size[1])
    accumulated_hist = { key: hist[key]/total for key in hist}
    
    last_pixel_value = -1
    for pixel_value in sorted(list(accumulated_hist.keys())):
        if last_pixel_value == -1:
            pass
        else:
            accumulated_hist[pixel_value] += accumulated_hist[last_pixel_value]
        last_pixel_value = pixel_value
    return accumulated_hist

def equalize_image(img):
    eq_hist = accumulated_histogram(img)
    
    def equalice_pixel(pixel_value):
        eq_pixel_value = eq_hist[pixel_value]
        return floor((eq_pixel_value - eq_hist[0])/(1 - eq_hist[0]) * 255 + 0.5)
    return img.point(lambda x: equalice_pixel(x))
```

```python
im = Image.open('img/ImagenesHistograma/Fig10.jpg')
hist = histogram(im)

eq_im = equalize_image(im)
eq_hist = histogram(eq_im)

### Ploteo imagen original
fig, (ax, ax2) = plt.subplots(ncols=2)

ax.imshow(im)
ax2.plot(list(hist.keys()), list(hist.values()))

ax.set_title("Imagen original")
ax2.set_title("Histograma original")

asp = np.abs(np.diff(ax2.get_xlim())[0] / np.diff(ax2.get_ylim())[0])
asp /= np.abs(np.diff(ax.get_xlim())[0] / np.diff(ax.get_ylim())[0])

ax2.set_aspect(asp)

plt.show()

### Ploteo imagen equalizada
fig, (ax, ax2) = plt.subplots(ncols=2)

ax.imshow(eq_im)
ax2.plot(list(eq_hist.keys()), list(eq_hist.values()))

ax.set_title("Imagen equalizada")
ax2.set_title("Histograma equalizado")

asp = np.abs(np.diff(ax2.get_xlim())[0] / np.diff(ax2.get_ylim())[0])
asp /= np.abs(np.diff(ax.get_xlim())[0] / np.diff(ax.get_ylim())[0])

ax2.set_aspect(asp)

plt.show()
```

#### 8) Aplicar la ecualización del histograma por segunda vez a la misma imagen. Obesrvar el resultado y dar una explicación de lo sucedido.

```python
eq2_im = equalize_image(eq_im)
eq2_hist = histogram(eq2_im)

### Ploteo imagen equalizada
fig, (ax, ax2) = plt.subplots(ncols=2)

ax.imshow(eq_im)
ax2.plot(list(eq_hist.keys()), list(eq_hist.values()))

ax.set_title("Imagen equalizada")
ax2.set_title("Histograma equalizado")

asp = np.abs(np.diff(ax2.get_xlim())[0] / np.diff(ax2.get_ylim())[0])
asp /= np.abs(np.diff(ax.get_xlim())[0] / np.diff(ax.get_ylim())[0])

ax2.set_aspect(asp)

plt.show()

### Ploteo imagen doblemente equalizada

fig, (ax, ax2) = plt.subplots(ncols=2)

ax.imshow(eq2_im)
ax2.plot(list(eq2_hist.keys()), list(eq2_hist.values()))

ax.set_title("Imagen doblemente equalizada")
ax2.set_title("Histograma doblemente equalizado")

asp = np.abs(np.diff(ax2.get_xlim())[0] / np.diff(ax2.get_ylim())[0])
asp /= np.abs(np.diff(ax.get_xlim())[0] / np.diff(ax.get_ylim())[0])

ax2.set_aspect(asp)

plt.show()

### Ploteo su diferencia
fig, (ax, ax2) = plt.subplots(ncols=2)

ax.set_title("Diferencias entre imagen equalizada y doblemente equalizada")
ax.imshow(substract(eq2_im, eq_im))

### Ploteo la funcion de densidad de la imagen equalizada
eq_h = accumulated_histogram(eq_im)

ax2.set_title("Función de densidad del valor de un pixel para la imagen equalizada")
ax2.bar(list(eq_h.keys()), list(eq_h.values()))

asp = np.abs(np.diff(ax2.get_xlim())[0] / np.diff(ax2.get_ylim())[0])
asp /= np.abs(np.diff(ax.get_xlim())[0] / np.diff(ax.get_ylim())[0])

ax2.set_aspect(asp)
plt.show()
```

Podemos observar que aplicar la equalización por segunda vez no genera ningún cambio con respecto a aplicarlo una sola vez. 

Si observamos la función de densidad del valor de un pixel de la imagen equalizada, podemos ver que es muy similar al de una variable con distribución aleatoria en el intervalo [0, 255]. Aplicarle el filtro de equalización a una imagen equalizada corresponde con intentar acercar su función de distribución aún más al de una variable aleatoria, pero esto no genera ningún efecto ya que al pasarle el filtro por primera vez obtuvimos el mejor resultado posible para la imagen dada.
